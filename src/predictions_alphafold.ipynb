````python
import pandas as pd
import requests
````

````python
dataset_file = "../data/output/dataset.tsv"
alphafold_dir = "../data/alphafold"

# Output
# Files in "CAID" format with UniProt accessions mapped to DisProt
````

````python
df_dataset = pd.read_csv(dataset_file, sep="\t")
df_dataset
````

````python
for method, file_name in zip(['AlphaFold-binding', 'AlphaFold-disorder', 'AlphaFold-rsa'], ['af_binding-25-0.581.dat', 'af_disorder.dat', 'af_disorder-25.dat']):
    with open("{}/{}".format(alphafold_dir, file_name)) as f:
        with open("{}/{}.caid".format(alphafold_dir, method), "w") as fout:
            for prot in f.read().split(">"):
                if prot:
                    lines = prot.strip().split("\n")
                    disprot_id = df_dataset.loc[df_dataset['acc'] == lines[0], 'disprot_id'].iloc[0]
                    fout.write(">{}\n{}\n".format(disprot_id, "\n".join(lines[1:])))
````

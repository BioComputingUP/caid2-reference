```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset_merge_file = "../data/output/dataset_merge.tsv"

# output
figures_dir = "../data/output/figures"
disprot_delta_file = "../data/output/disprot_delta.tsv"
```

```python
df_dataset = pd.read_csv(dataset_merge_file, sep="\t", header=[0,1], index_col=[0,1,2,3,4])
df_dataset
```

```python
df_dataset['disorder-mean'] = df_dataset.loc[:, ('disorder', )].mean(axis=1)
df_dataset['binding-mean'] = df_dataset.loc[:, ('binding', )].mean(axis=1)
df_dataset
```

```python
# Modify reference values for the figure offset
df_dataset['af-plddt'] = 1 - df_dataset[('alphafold', 'af-disorder')]
df_dataset.loc[df_dataset[('reference', 'disorder')] == 1, 'disorder-y'] = 1.3
df_dataset.loc[df_dataset[('reference', 'binding')] == 1, 'binding-y'] = 1.2
df_dataset.loc[df_dataset[('reference', 'pdb')] == 1, 'pdb'] = 1.1
df_dataset
```

```python
df_ = df_dataset.groupby(['disprot_id', 'acc']).apply(lambda x: ((x['disorder-y'].sum() - x[('alphafold', 'af-rsa')].sum()) * 100 / x.index.get_level_values(3).max(), (x['binding-y'].sum() - x[('binding-mean', '')].sum()) * 100 / x.index.get_level_values(3).max())).reset_index(name='delta')
df_[['delta-disorder', 'delta-binding']] = pd.DataFrame(df_.delta.tolist(), index= df_.index)
df_.drop(columns='delta', inplace=True)
df_.sort_values(['delta-disorder', 'delta-binding'], inplace=True)
df_.to_csv(disprot_delta_file, sep="\t", index=False)
df_
```

```python
# Generate all the figures
for name, df_g in df_dataset.groupby(['disprot_id', 'acc']):
    df_g = df_g.loc[:, [('disorder-y', ''), ('binding-y', ''), ('pdb', ''), ('binding-mean', ''), ('disorder-mean', ''), ('alphafold', 'af-rsa')]]
    df_g.reset_index(level='pos', inplace=True)
#     print(df_g)
#     break
    
    ax = df_g.plot(x='pos', ylim=(0,1.4), grid=True, figsize=(30,8), fontsize=18)
    ax.set_title(" - ".join(name), fontdict={'fontsize':18})
    ax.xaxis.label.set_visible(False)
    style=['-', '-', '-', '-', '-', '-']
    lw=[6, 6, 6, 3, 2, 3]
    for i, l in enumerate(ax.lines):
        plt.setp(l, linestyle=style[i], linewidth=lw[i])
    ax.hlines(0.5, xmin=1, xmax=df_g['pos'].max(), lw=1, linestyle='-', color='red')
    ax.set_xlim(1, df_g['pos'].max())
    ax.legend(prop={'size': 18}, loc='center left', bbox_to_anchor=(1.0, 0.5))
    fig = ax.get_figure()
    fig.savefig("{}/{}.png".format(figures_dir, "_".join(name)), bbox_inches='tight')
    plt.close(fig)
    
```

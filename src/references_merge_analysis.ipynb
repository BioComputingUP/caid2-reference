```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset_merge_file = "../data/output/dataset_merge.tsv"

# output
figures_dir = "../data/output/references_merge_analysis"
disprot_delta_file = "../data/output/disprot_delta.tsv"
```

```python
df_dataset = pd.read_csv(dataset_merge_file, sep="\t", header=[0,1], index_col=[0,1,2,3,4])
df_dataset
```

```python
df_dataset['disorder-mean'] = df_dataset.loc[:, ('disorder', )].mean(axis=1)
df_dataset['binding-mean'] = df_dataset.loc[:, ('binding', )].mean(axis=1)
df_dataset
```

```python
# Modify reference values for the figure offset
# df_dataset['af-plddt'] = 1 - df_dataset[('alphafold', 'af-disorder')]
df_dataset.loc[df_dataset[('reference', 'disorder')] == 1, 'disorder-y'] = 1.15
df_dataset.loc[df_dataset[('reference', 'binding')] == 1, 'binding-y'] = 1.1
df_dataset.loc[df_dataset[('reference', 'pdb')] == 1, 'pdb'] = 1.05
df_dataset
```

```python
# Generate all the figures
"""
orange (1.0, 0.4980392156862745, 0.054901960784313725)
red (0.8392156862745098, 0.15294117647058825, 0.1568627450980392)
blue (0.12156862745098039, 0.4666666666666667, 0.7058823529411765)
green (0.17254901960784313, 0.6274509803921569, 0.17254901960784313)
violet (0.5490196078431373, 0.33725490196078434, 0.29411764705882354)
violet2 (0.8901960784313725, 0.4666666666666667, 0.7607843137254902)
gray (0.4980392156862745, 0.4980392156862745, 0.4980392156862745)
yellow (0.7372549019607844, 0.7411764705882353, 0.13333333333333333)
cyan (0.09019607843137255, 0.7450980392156863, 0.8117647058823529)
"""
colors_reference = [(1.0, 0.4980392156862745, 0.054901960784313725),
            (0.8392156862745098, 0.15294117647058825, 0.1568627450980392),
            (0.12156862745098039, 0.4666666666666667, 0.7058823529411765)]

colors_pred = [(1.0, 0.4980392156862745, 0.054901960784313725),
            (0.8392156862745098, 0.15294117647058825, 0.1568627450980392),
            (0.4980392156862745, 0.4980392156862745, 0.4980392156862745)]

for name, df_g in df_dataset.groupby(['disprot_id', 'acc']):
    df_g.reset_index(level='pos', inplace=True)

    fig, axes = plt.subplots(2,1, figsize=(30,10), gridspec_kw={'height_ratios': [2, 6]})
    plt.subplots_adjust(wspace=0, hspace=0)

    # Reference
    df_g.loc[:, [('pos', ''), ('disorder-y', ''), ('binding-y', ''), ('pdb', '')]].plot(
        x='pos', ylim=(0,1.4), grid=True, ax=axes[0], fontsize=18, color=colors_reference, linewidth=6)

    # axes[1].hlines(0.331, xmin=1, xmax=df_g['pos'].max(), lw=1, linestyle='--', color= (0.4980392156862745, 0.4980392156862745, 0.4980392156862745))
    axes[1].hlines(0.5, xmin=1, xmax=df_g['pos'].max(), lw=1, linestyle='-', color='green')

    # Prediction
    df_g.loc[:, [('pos', ''), ('disorder-mean', ''), ('binding-mean', ''), ('disorder', 'AlphaFold-pLDDT')]].plot(
        x='pos', ylim=(0,1.4), grid=True, ax=axes[1], fontsize=18, color=colors_pred, linewidth=3, alpha=0.5)

   

    axes[0].set_title(" - ".join(name), fontdict={'fontsize':18})
    axes[0].set_ylim(1, 1.4)
    axes[0].set_xlim(1, df_g['pos'].max())
    axes[0].axis('off')
    h,_ = axes[0].get_legend_handles_labels()
    axes[0].legend(h, ['Disorder-DisProt', 'Binding-DisProt', 'PDB observed'],
                   prop={'size': 18}, loc='lower left', bbox_to_anchor=(-0.007, 0.5), ncol=3)

    axes[1].set_ylim(0, 1.1)
    axes[1].set_xlim(1, df_g['pos'].max())
    axes[1].xaxis.label.set_visible(False)
    # axes[1].set_yticks(np.arange(0,1.1,0.1))
    axes[1].set_yticks([0, 0.5, 1.0])
    h,_ = axes[1].get_legend_handles_labels()
    axes[1].legend(h, ['Disorder pred. average', 'Binding pred. average', 'AlphaFold-pLDDT'],
                   prop={'size': 18}, loc='lower left', bbox_to_anchor=(0.51, 1.17), ncol=3)

    fig.savefig("{}/{}.png".format(figures_dir, "_".join(name)), bbox_inches='tight', dpi=300)
    plt.close(fig)
    # break
```

```python
# Find cases where Disprot strongly disagree with AlphaFold
df_ = df_dataset.groupby(['disprot_id', 'acc']).apply(lambda x: ((x['disorder-y'].sum() - x[('disorder', 'AlphaFold-pLDDT')].sum()) * 100 / x.index.get_level_values(3).max(), (x['binding-y'].sum() - x[('binding-mean', '')].sum()) * 100 / x.index.get_level_values(3).max())).reset_index(name='delta')
df_[['delta-disorder', 'delta-binding']] = pd.DataFrame(df_.delta.tolist(), index= df_.index)
df_.drop(columns='delta', inplace=True)
df_.sort_values(['delta-disorder', 'delta-binding'], inplace=True)
df_.to_csv(disprot_delta_file, sep="\t", index=False)
df_
```
